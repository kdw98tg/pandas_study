# TimeSeries Index


- pd.to_datetime
- pd.date_range
- year,month,day 추출하는 방법
- 월별 수익률을 구하는 방법


```python
import pandas as pd

ten_year_rate_path = "/Users/kimdongwoo/pandas_study/pandas/강의자료/Data/10Y Rate.csv"
df = pd.read_csv(ten_year_rate_path, index_col = "Date")
df.index = pd.to_datetime(df.index)
df
```

```python
# 인덱스를 만드는 과정임
# 2020-10-26 부터 끝은 지정하지 않고, 100개의 날짜를 생성함
timeSeries_index = pd.date_range(start="2020-10-26", end = None, periods = 100) 
timeSeries_index
```

```python
df.index.year
df.index.month
df.index.day

df["Year"] = df.index.year
df["Month"] = df.index.month
df["Day"] = df.index.day
```

```python
import numpy as np
df['Condition'] = np.where(df['Month'] == df['Month'].shift(1),0,1)
df["IndexNumber"] = range(df.shape[0])

index_condition_number = df[df["Condition"] == 1]["IndexNumber"].values - 1
df_condition = df.iloc[index_condition_number[1:]]
df_monthly_return = df_condition["Close"].pct_change()
df_monthly_return
```
