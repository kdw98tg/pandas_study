# Line Plot


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

kospi_path = "/Users/kimdongwoo/pandas_study/pandas/강의자료/Data/KOSPI.csv"
df_kospi = pd.read_csv(kospi_path, index_col = "Date")
df_kospi
```

```python
df_kospi.plot()
```

```python
plt.figure() # 도화지를 생성

df_kospi.plot(figsize=(15,4)) # x축으로 15만큼, y축으로 4만큼의 크기를 가진 차트를 생성함

plt.axhline(2000, linestyle = "--", color="red") # 선긋기

plt.show()
```

```python
fig, ax = plt.subplots()

df_kospi.plot(kind = 'line', ax = ax, figsize = (15,10)) # df_kospi의 데이터를 선 그래프로 ax 에 그림

ax.axhline(2000, linestyle = '--', color = 'green') # 선그리기

ax.set(xlabel = "Date", ylabel = 'KOSPI', title = 'Time Series of Kospi Index') # x축 이름 : Date, y축 이름 : KOSPI, 그래프 제목 : Time Series ...

plt.show()
```

# 두개 이상의 그래프


```python
samsung_naver_path = "/Users/kimdongwoo/pandas_study/pandas/강의자료/Data/samsung_naver_excel.xlsx"
df_samsung = pd.read_excel(samsung_naver_path, sheet_name=0, index_col = "Date")
df_samsung
```

```python
fig, axs= plt.subplots(2, 1, figsize= (15,10))


axs[0].plot(df_samsung["Close"]["2018-05-04":],color="blue") # 2018년부터 Close 라는 열의 데이터를 나타냄
axs[0].grid(True) # 눈금 그어줌
axs[0].set(xlabel="Date",ylabel="Price",title="Time Series of Samsung Price") # 제목이랑 라벨 설정
axs[0].plot(df_samsung["Open"]["2018-05-04":], color = "red") #Open 열의 2018-05-04 부터의 데이터를 나타냄

axs[1].scatter(df_samsung.index, df_samsung['Volume'], color = 'red') # 산점도 그래프를 그림
axs[1].grid(True)
axs[1].set(xlabel = "Date", ylabel = 'Volume')
```

```python
#plt 의 그래프를 전역으로 설정함
#plt.rcParams["figure.figsize"] = (22,15)
#plt.rc("font", family="Malgun Gothic")
#plt.rcParams["font.size"] = 15
plt.rc("axes", unicode_minus=False)

samsung_kospi_path = "/Users/kimdongwoo/pandas_study/pandas/강의자료/Data/samsung_kospi_excel.xlsx"
df=pd.read_excel(samsung_kospi_path, sheet_name=0, index_col="Date")
df= df.loc["2018-01-02":]
df
```

```python
df.index[df['Close'].argmax()]
```

```python
fig, axes = plt.subplots(3,1, sharex = False)

axes[0].plot(df['Close'], color = 'red', linewidth = 2)
axes[0].plot(df['Close'].rolling(window = 30).mean(), color = 'blue', linewidth = 3)

#이동평균선 그리기
axes[0].axhline(df['Close'].mean(), linestyle = '--', linewidth = 5)

# 수평점선 : 전체 평균 종가
axes[0].scatter(df.index[df['Close'].argmax()], df['Close'].max(), marker = 'o',s = 500)

#최고가에 원 표시
axes[0].legend(['Close', 'Rolling Mean 30days'])
axes[0].set_title("Samsung Stock")

# 일간 수익률
axes[1].plot(df['Change']*100, color ='blue', linewidth = 0.8)
axes[1].axhline(0, linestyle = '--', linewidth = 4, color = 'red')
axes[1].tick_params(axis = 'x', labelrotation = 45)
axes[1].set_title("수익률")

# 거래량
axes[2].fill_between(df.index, df['Volume'], color = 'green', alpha = 1)
axes[2].set_title("거래량")
axes[2].axvline(df.index[df['Volume'].argmax()], color = 'red', linewidth = 5, linestyle = '--')

plt.tight_layout() # 여백 자동 조정 → 라벨이나 제목이 겹치지 않도록
plt.show()
```
